```python
import matplotlib.pyplot as plt

value_lst_max = []
value_lst_avg = []

for i in range(8):
    value_max = 0
    value_avg = 0
    for j in range(100):
        model = simulation.NKModel(N=8, K=i, R=3, G=2, U=0)
        max_key = max(model.landscape, key=model.landscape.get)
        max_value = model.landscape[max_key]
        value_max += max_value
        value_avg += sum(model.landscape.values())/len(model.landscape)
    value_lst_max.append(value_max/100)
    value_lst_avg.append(value_avg/100)

plt.plot(range(8), value_lst_max, label="Max")
plt.plot(range(8), value_lst_avg, label="Average")
plt.legend()
```
